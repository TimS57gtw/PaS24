# Physik Am Samstag

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
data_x = np.linspace(0, 10, 11)
data_y = 0.2 * data_x + np.random.normal(1, 0.4, data_x.shape)
```

```python
ideal_x = np.linspace(0, 10, 11)
ideal_y = 0.2 * data_x + 1
lin_xs = np.linspace(-0.2, 10.2, 20)
```

## Erwartete Messergebnisse

```python
plt.title('Ideale Ergebnisse')
plt.plot(ideal_x, ideal_y, c='red', ls='--')
plt.scatter(ideal_x, ideal_y, marker='x', s=200, c='black')
plt.grid()
plt.xlabel('Spannung U (V)')
plt.ylabel('Stromstärke I (A)')
plt.xlim(-1, 11)
# plt.axhline(y=0, c='black')
plt.ylim(-0.2, 3.5)
# plt.axvline(x=0, c='black')
```

```python
plt.title('Echte Ergebnisse')
# plt.plot(ideal_x, ideal_y, c='red', ls='--')
plt.scatter(data_x, data_y, marker='x', c='black', s=200)
plt.grid()
plt.xlabel('Spannung U (V)')
plt.ylabel('Stromstärke I (A)')
plt.xlim(-1, 11)
# plt.axhline(y=0, c='black')
plt.ylim(-0.2, 3.5)
# plt.axvline(x=0, c='black')
```

## Lösung 1: Reression per Hand

```python
m = 0.3
b = 0.5
ys = m * lin_xs + b 
```

```python
plt.plot(lin_xs, ys, c='red', lw=3)
plt.title('Manueller Fit')
# plt.plot(ideal_x, ideal_y, c='red', ls='--')
plt.scatter(data_x, data_y, marker='x', c='black', s=200)
plt.grid()
plt.xlabel('Spannung U (V)')
plt.ylabel('Stromstärke I (A)')
plt.xlim(-1, 11)
# plt.axhline(y=0, c='black')
# plt.ylim(-0.2, 3.5)
# plt.axvline(x=0, c='black')
```
